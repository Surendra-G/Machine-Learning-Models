## Splitting the Data into train set and Test set

```python
from sklearn.model_selection import train_test_split  #Importing the train_test_split method from the  sklearn library

X_train, X_test,y_train, y_test = train_test_split(x, y, test_size=0.3, random_state = 42)
```

## Building Two Machine LEarnig Model
1.   Linear Regression




```python
from sklearn.linear_model import LinearRegression  #Importing the LinearRegression method from the  sklearn library

# Model 1: Linear Regression
linearRegressionModel = LinearRegression()
linearRegressionModel.fit(X_train, y_train)
```

2. RandomForest Regression

```python
from sklearn.ensemble import RandomForestRegressor   #Importing the RandomForestRegressor method from the  sklearn library
RandomforestModel = RandomForestRegressor()
RandomforestModel.fit(X_train, y_train)
```

## Evaluating both model on Test Dataset.

```python
from sklearn.metrics import mean_squared_error   #Importing the mean_squared_error & r2_score method from the  sklearn library
from sklearn.metrics import  r2_score

# Prediction on the test set
y_pred_linearRegression = linearRegressionModel.predict(X_test)
y_pred_randomForest = RandomforestModel.predict(X_test)

# Evaluating the Linear Regression model by mean square error method and r2_score methods
mse_for_linearRegression = mean_squared_error(y_test, y_pred_linearRegression)
r2_for_linearRegression = r2_score(y_test, y_pred_linearRegression)

# Evaluating the Random Forest model by mean square error method and r2_score methods
mse_for_Randomforest = mean_squared_error(y_test, y_pred_randomForest)
r2_for_Randomforest = r2_score(y_test, y_pred_randomForest)
```

## Displaying the results of Mean Square Error and R-squared of Both Linear Regression and Random FOrest



```python
#Displaying all the Terms
print("Linear Regression:")
print(f"Mean Squared Error: {mse_for_linearRegression}")
print(f"R-squared: {r2_for_linearRegression}")
print("\nRandom Forest Regression:")
print(f"Mean Squared Error: {mse_for_Randomforest}")
print(f"R-squared: {r2_for_Randomforest}")
```

According to this result, Random Forest Regression provides us significantly less Mean Square Error and high R squared in comparison with Linear Regression which means Random forest provides us model's prediction close to actual value and can able to include or capture the varience in the data in comparison with linear regression


## Hyper Parameter Optimization and cross Validation

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.feature_selection import RFE

# Define parameter grid
param_grid = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Initialize RandomizedSearchCV
random_search = RandomizedSearchCV(RandomForestRegressor(),
                                   param_distributions=param_grid,
                                   n_iter=50,
                                   cv=5,
                                   scoring='neg_mean_squared_error',
                                   n_jobs=-1,
                                   random_state=42)

# Fit the randomized search
random_search.fit(X_train, y_train)

# Get the best parameters and model
best_params_rf = random_search.best_params_
best_model_rf = random_search.best_estimator_

# Evaluate the best model on test data
Y_pred = best_model_rf.predict(X_test)
mse = mean_squared_error(y_test, Y_pred)
r2 = r2_score(y_test, Y_pred)

# Print results
print("Best Hyperparameters:")
print(best_params_rf)
print("\nBest Model:")
print(best_model_rf)
print("\nPerformance Metrics:")
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

In this code, the dataset is restricted to the best features selected using RFE, including 'male', 'age', 'cigsPerDay', 'sysBP', 'glucose'. After that, the data is splitted into training and testing sets, followed by standardization of features. The logistic regression model is initialized with the best hyperparameters obtained from hyperparameter optimization. Finally, the model is trained on the training data and used to make predictions on both the training and testing sets.

```python
# Define the best features selected using RFE
X = new_df[['male', 'age', 'cigsPerDay', 'sysBP', 'glucose']]
Y = new_df['TenYearCHD']

# Split the data into training and testing sets
new_X_train, new_X_test,new_y_train, new_y_test = train_test_split(X, Y, test_size=0.3, random_state = 42)

# Standardize features by removing the mean and scaling to unit variance
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(new_X_train)
X_test_scaled = scaler.transform(new_X_test)

# Create logistic regression model with best hyperparameters
logistic_model = LogisticRegression(C=10, penalty='l2')

# Fit the model on the training data
logistic_model.fit(X_train_scaled, new_y_train)

# Make predictions on the test set
y_test_pred = logistic_model.predict(X_test_scaled)

# Make predictions on the train set
y_train_pred = logistic_model.predict(X_train_scaled)
```

The below codes computes the accuracy of the model on both the training and testing datasets using the accuracy_score function. Additionally, it generates classification reports using the classification_report function, which includes precision, recall, F1-score, and support for each class. Finally, it prints out the accuracy and classification reports for both the training and testing datasets.

```python
# Evaluate the model
accuracy_train = accuracy_score(new_y_train, y_train_pred)
accuracy_test = accuracy_score(new_y_test, y_test_pred)

report_train = classification_report(new_y_train, y_train_pred)
report_test = classification_report(new_y_test, y_test_pred)

print(f'Train Accuracy: {accuracy_train}')
print('\nTrain Classification Report:\n', report_train)

print(f'\nTest Accuracy: {accuracy_test}')
print('\nTest Classification Report:\n', report_test)
```


According to the above result, the  model shows the strong performance in prediction on class 0 but struggling with class 1.It also gives the decent overall accuracy, in  precision, recall, and F1-score for class 0 are notably lower compared to class 1.
