```python
#Data Cleaning

#filling the missing values
#By observing my Daatsets, i only found float64 & object as datatype.
#For float64, i used mean of the columns for filling out the missing value
numeric_columns = df.select_dtypes(include=['float64']).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

```

```python
# For object datatype, i use mode of the columns for filling the missing space.
categorical_columns = df.select_dtypes(include=['object']).columns
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])

#Displaying the data after filling out the missing values
df.isnull().sum()
```

```python
# droping the rows having missing values.
df = df.dropna()

# Removing duplicates
df = df.drop_duplicates()
df
```

```python
# Reset index after dropping rows
df = df.reset_index(drop=True)
```

```python
# Display the first few rows of the cleaned DataFrame
print(df.head())
```

```python
#Summary Satistics

#Question 1:
# Spliting the Dataset into numeric and categorical columns
numeric_columns = df.select_dtypes(include=['float64']).columns
categorical_columns = df.select_dtypes(include=['object']).columns

# For Numeric Columns
numeric_summary = df[numeric_columns].describe()
print("\nSummary Statistics for Numeric Columns:")
print(numeric_summary)


#unique values for numeric column
for col in numeric_columns:
  numeric_unique_values = df[col].unique()
  print(f"\n Numeric Unique values for \n {col}: {numeric_unique_values}")

```

```python
# For Categorical Columns
#Question 3
categorical_summary = df[categorical_columns].describe(include='all')
print("\nSummary Statistics for Categorical Columns:")
print(categorical_summary)

```

```python
# Unique values for character variables
for col in categorical_columns:
    unique_values = df[col].unique()
    print(f"\nUnique values for {col}: {unique_values}")
```

```python
#Question 1
# Calculating total goals for home and away teams
total_home_goals = df['Home Team Goals'].sum()
print("Total Home Team Goal: ",total_home_goals)
total_away_goals = df['Away Team Goals'].sum()
print("Total Away Team Goal: ",total_away_goals)
print("\n\n")

# Creating a bar chart
plt.figure(figsize=(8, 6))
bar1 = plt.bar('Home Team Goals', total_home_goals, color='blue', label='Home Team Goals')
bar2 = plt.bar('Away Team Goals', total_away_goals, color='green', label='Away Team Goals')
plt.title('Total Goals Scored by Home and Away Teams')
plt.xlabel('Team')
plt.ylabel('Total Goals')
plt.savefig("GoalScored.png")
plt.legend()
plt.show()

```

```python
# Creating distribution plots
plt.figure(figsize=(12, 5))

# Distribution plot for Home Team Goals
plt.subplot(1, 2, 1)
sns.histplot(df['Home Team Goals'],bins=10,color="red",kde=True,line_kws= {'linewidth': '2'}).lines[0].set_color('green')
plt.title('Distribution of Home Team Goals')
plt.xlabel('Goals')
plt.ylabel('Frequency')

#skewness for Home Team Goals
home_goals_skew = df['Home Team Goals'].skew()
print(f"The skewness of the Home team goals is:{home_goals_skew} ")


# Distribution plot for Away Team Goals
plt.subplot(1, 2, 2)
sns.histplot(df['Away Team Goals'],bins=10,color='red',kde=True,line_kws= {'linewidth': '2'}).lines[0].set_color('green')
plt.title('Distribution of Away Team Goals')
plt.xlabel('Goals')
plt.ylabel('Frequency')

#skewness for Away Team Goals
away_goals_skew = df['Away Team Goals'].skew()
print(f"The skewness of the Away team goals is: {away_goals_skew}\n\n")

plt.tight_layout()
plt.savefig("histogram.png")
plt.show()
```

```python
# Creating boxplots to identify outliers
plt.figure(figsize=(8, 6))

# Boxplot for Home Team Goals
plt.subplot(1, 2, 1)
boxprops_home = dict(color='blue', linewidth=3)
plt.boxplot(df['Home Team Goals'], boxprops=boxprops_home, medianprops=dict(color='black'))
plt.title('Boxplot of Home Team Goals')
plt.grid()

# Boxplot for Away Team Goals
plt.subplot(1, 2, 2)
boxprops_away = dict(color='green', linewidth=3)
plt.boxplot(df['Away Team Goals'], boxprops=boxprops_away, medianprops=dict(color='black'))
plt.title('Boxplot of Away Team Goals')
plt.savefig("outliers.png")
plt.grid()

plt.tight_layout()
plt.show()
```
